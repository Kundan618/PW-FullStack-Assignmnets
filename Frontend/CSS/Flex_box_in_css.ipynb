**1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When would you choose to use one over the other?**

#CSS Flexbox#

Flexbox, or the Flexible Box Layout, is a one-dimensional layout model in CSS. It is used to arrange items in a single direction, either as a row or a column. Flexbox is content-driven, which means the layout adjusts based on the size and number of items within the container. It provides powerful tools for alignment, spacing, and ordering of items, making it ideal for creating responsive elements like navigation bars, buttons, or lists. Flexbox is best used when the layout needs to be flexible in one direction only, and it helps reduce the need for floats and manual positioning.


**Use Flexbox when:**

* You have a linear layout (e.g., navigation bars, toolbars, lists).
* Items should wrap or stretch depending on screen size.
* You need simple alignment and distribution of space.
* The layout is content-driven and doesn’t need a strict grid structure.

**Example use cases:**

* Horizontal nav menu
* Centering a button within a container
* Aligning form elements in a row

#CSS Grid#

CSS Grid is a two-dimensional layout system that allows developers to design web layouts using both rows and columns. It is layout-driven, which means you define the structure first—by creating a grid of cells—and then place items within those cells. Grid is especially useful for complex and consistent layouts, like web pages with headers, sidebars, main content, and footers. It provides precise control over element positioning and supports overlapping elements and grid gaps. CSS Grid is ideal when you need to build structured, grid-based layouts across two axes.



**Use Grid when:**

* You’re creating entire page layouts or components with rows and columns.
* You want precise control over both dimensions of the layout.
* The layout structure is more complex or consistent (e.g., templates).
* You need overlapping elements or layering within a grid.

**Example use cases:**

* Website layout with header, sidebar, main, and footer
* Image galleries
* Dashboard interfaces





**2. Explain the role of the following key properties in the Flexbox layout mode**
 1. justify-content
 2. align-items
 3. gap
 4. flex-direction
 5. flex-wrap


**Justify Content**
* Aligns the entire flex container's items along the main axis.
* Possible values: flex-start, flex-end, center, space-between, space-around,  space-evenly.
* Determines how space is distributed around items.

 **Align Item**

* Aligns flex items along the cross-axis.
* Possible values: stretch, center, flex-start, flex-end.
* The default value is stretch.

**Gap**

* Sets the spacing between items in a flex container.
* Can be defined for rows (row-gap) and columns (column-gap), or both (gap).
* Useful for creating consistent spacing without margins.


**Flex Direction**
* Specifies the direction of the flex items in the flex container.
* Possible values: row, row-reverse, column, column-reverse.
* Determines the main axis direction (horizontal or vertical).

**Flex Wrap**
* Controls whether flex items are forced into a single line or can wrap onto multiple lines.
* Possible values: nowrap, wrap, wrap-reverse.
* The default value is nowrap.
